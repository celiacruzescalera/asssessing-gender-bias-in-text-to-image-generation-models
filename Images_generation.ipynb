```python
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
import torch
```

# MODEL 1

```python
# Load Model 1: runwayml/stable-diffusion-v1-5

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")
```

### GENDER NEUTRAL IMAGES

```python
# Get the professions gender neutral prompts
with open("./Prompts/PROMPTS_neutral_professions.txt", "r") as f:
    professions_NO_gender = f.read().splitlines()

# Get the COCO gender neutral prompts
with open("./Prompts/PROMPTS_neutral_coco.txt", "r") as f:
    COCO_NO_gender = f.read().splitlines()

no_gender = professions_NO_gender + COCO_NO_gender
```

```python
# Generate the gender neutral images
k = 1
for prompt in no_gender:
  images = pipe(prompt, num_images_per_prompt=10).images
  if k < 10:
    num_linia = '0' + str(k)
  else:
    num_linia = str(k)
  for i in range(10):
    if (i + 1) < 10:
      id = "01_" + num_linia + "_0" + str(i+1) + "_nogender.png"
    else:
      id = "01_" + num_linia + "_" + str(i+1) + "_nogender.png"
    images[i].save("./Images/Model_1/Gender_Neutral/" + id)
  k += 1
```

### GENDER SPECIFIC IMAGES



```python
# Get the professions gender specific prompts
with open("./Prompts/PROMPTS_gender_professions.txt", "r") as f:
    # Read the contents of the file and split it into lines
    professions_gender = f.read().splitlines()

# Get the COCO gender specific prompts
with open("./Prompts/PROMPTS_gender_coco.txt", "r") as f:
    # Read the contents of the file and split it into lines
    COCO_gender = f.read().splitlines()

gender = professions_gender + COCO_gender
```

```python
# Generate the gender specific images
k = 1
for prompt in gender:
  images = pipe(prompt, num_images_per_prompt=10).images
  if k < 10:
    num_linia = '0' + str(k)
  else:
    num_linia = str(k)
  if k % 2 == 0:
    for i in range(10):
      if (i + 1) < 10:
        id = "02_" + num_linia + "_0" + str(i+1) + "_femenine.png"
      else:
        id = "02_" + num_linia + "_" + str(i+1) + "_femenine.png"
      images[i].save("./Images/Model_1/Gende_Specific/" + id)
  else:
    for i in range(10):
      if (i + 1) < 10:
        id = "02_" + num_linia + "_0" + str(i+1) + "_masculine.png"
      else:
        id = "02_" + num_linia + "_" + str(i+1) + "_masculine.png"
      images[i].save("./Images/Model_1/Gender_Specific/" + id)
  k += 1
```
