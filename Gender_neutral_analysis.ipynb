```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

# MODEL 1

### MANUAL

```python
# Manual classificaction model 1
with open("./Model1_manual_classification.json", "r") as file:
    data = json.load(file)
```

```python
# Global count

man_manual1 = 0
woman_manual1 = 0
dk_manual1 = 0
none_manual1 = 0 

for key in data:
  value = data[key]
  if (value == "man"):
    man_manual1 += 1
  elif (value == "woman"):
    woman_manual1 += 1
  elif (value == "dk"):
    dk_manual1 += 1
  elif (value == "none"):
    none_manual1 += 1
  elif (value == 'both'):
    man_manual1 += 1
    woman_manual1 += 1
print("Number of 'Masculine' (manual model 1): " + str(man_manual1))
print("Number of 'Feminine' (manual model 1): " + str(woman_manual1))
print("Number of 'Unidentifiable' (manual model 1): " + str(dk_manual1 + none_manual1))
```

```python
# Pie chart 
results = [386,190,143]
names = ["Masculine","Feminine","Unidentifiable"]
colors = ["palegreen", "mediumseagreen", "seagreen"]
plt.pie(results, textprops={'fontsize': 15}, labels=names, colors=colors, autopct="%0.1f %%")
plt.show()
```

```python
# Pie chart without 'Unidentifiable'
results = [386,190]
names = ["Masculine","Feminine"]
colors = ["palegreen", "mediumseagreen"]
plt.pie(results, textprops={'fontsize': 15}, labels=names, colors=colors, autopct="%0.1f %%")
plt.show()
```

```python
# Specific count

man_50_60 = 0
man_60_70 = 0
man_70_80 = 0
man_80_90 = 0
man_90_100 = 0

woman_50_60 = 0
woman_60_70 = 0
woman_70_80 = 0
woman_80_90 = 0
woman_90_100 = 0

percentatges_50_60 = 0
percentatges_60_70 = 0
percentatges_70_80 = 0
percentatges_80_90 = 0
percentatges_90_100 = 0

for i in range(71):
  if((i + 1) < 10):
    num_img_principi = "01_0" + str(i + 1)
  else: 
    num_img_principi = "01_" + str(i + 1)
  
  man = 0
  woman = 0

  for j in range(10):
    num_img = num_img_principi
    if((j + 1) < 10):
      num_img += "_0" + str(j + 1)
    else: 
      num_img += "_" + str(j + 1)
    
    value = data[num_img]
    if (value == "man"):
      man += 1
    elif (value == "woman"):
      woman += 1
    elif (value == "both"):
      man += 1
      woman += 1
  
  total = man + woman
  if (man == 0):
    percentatge_man = 0
  else:
    percentatge_man = (man*100)/total
  if (woman == 0):
    percentatge_woman = 0
  else:
    percentatge_woman = (woman*100)/total

  if (man > woman):
    percentatge = percentatge_man
    genere = "man"
  else:
    percentatge = percentatge_woman
    genere = "woman"

  if (percentatge >= 50 and percentatge <= 60):
    percentatges_50_60 += 1
    if (genere == "man"):
      man_50_60 += 1
    else:
      woman_50_60 += 1
  elif (percentatge > 60 and percentatge <= 70):
    percentatges_60_70 += 1
    if (genere == "man"):
      man_60_70 += 1
    else:
      woman_60_70 += 1
  elif (percentatge > 70 and percentatge <= 80):
    percentatges_70_80 += 1
    if (genere == "man"):
      man_70_80 += 1
    else:
      woman_70_80 += 1
  elif (percentatge > 80 and percentatge <= 90):
    percentatges_80_90 += 1
    if (genere == "man"):
      man_80_90 += 1
    else:
      woman_80_90 += 1
  elif (percentatge > 90 and percentatge <= 100):
    percentatges_90_100 += 1
    if (genere == "man"):
      man_90_100 += 1
    else:
      woman_90_100 += 1

print("Number of percentages between 50% and 60%: " + str(percentatges_50_60))
print("Masculine: " + str(man_50_60))
print("Feminine: " + str(woman_50_60))
print("Number of percentages between 60% and 70%: " + str(percentatges_60_70))
print("Masculine: " + str(man_60_70))
print("Feminine: " + str(woman_60_70))
print("Number of percentages between 70% and 80%: " + str(percentatges_70_80))
print("Masculine: " + str(man_70_80))
print("Feminine: " + str(woman_70_80))
print("Number of percentages between 80% and 90%: " + str(percentatges_80_90))
print("Masculine: " + str(man_80_90))
print("Feminine: " + str(woman_80_90))
print("Number of percentages between 90% and 100%: " + str(percentatges_90_100))
print("Masculine: " + str(man_90_100))
print("Feminine: " + str(woman_90_100))
```
